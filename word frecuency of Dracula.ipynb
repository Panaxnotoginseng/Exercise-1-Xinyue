##### First,I get the text from the file Dracula.txt

```python
x = open('Dracula.txt', 'r', encoding = 'utf-8').read() 
```

##### In order to avoid counting the same word as two words because of case differences, I transfer all words into lower class.

```python
x = x.lower()  
```

##### After that, I replace everything that is not a word (punctuations and numbers) by space. 
##### Then I split text into seperated words by space and these seperated words form a list.

```python
import re
x = re.sub(r'[^a-zA-Z]',' ',x)
x_list = x.split(" ")  
len(x_list)
```

##### Now I create an empty dictionary and start to put words into it.

##### For each word, I check if it exists as a key in the dictionary. If it does, then add 1 to its value, if it doesn't, then this word will be added to the dictionary as a new key with value 1.

##### During the process, a key that haves no content at all would be added to the dictionary. Here I delete it.

```python
bookdict={}
for word in x_list:  
    key = word 
    if key in bookdict:  
        i = bookdict.get(key)
        i +=1
        bookdict[key] = i
    else:  
        bookdict.update({key:1})
del bookdict[""]
len(bookdict)
```

##### After adding all words to the dictionary, I import itemgetter to sort the order of the items by their values, from the biggest to the smallest. Normally it would sort from the smallest to the biggest, so I have to give truth value to "reverse".

##### Finally I print all the keys and their values using a loop, so that only one word and its frecuency will be presented in each line. It would be easier to read in this way.

#### EXTRA QUESTION: 
##### The result should roughly fits Zipf's law, which holds, approximately, when a list of measured values is sorted in decreasing order. It states that the value of the nth entry is inversely proportional to n. It is usually found that the most common word occurs approximately twice as often as the next common one, three times as often as the third most common, and so on. The result I get here does not precisely reflect the law, since the corpus is too small.

```python
import operator
from operator import itemgetter
sort_dict= dict(sorted(bookdict.items(), key=operator.itemgetter(1), reverse = True)) 
for key, value in sort_dict.items():
    print(key, value)  
```

```python
import matplotlib.pyplot as plt
import numpy as np
A = list (sort_dict.items ()) [:100]
B = list () 
C = list () 
for key, value in A:
    B.append(key)
    C.append(value)

x_axis_data = B
y_axis_data = C

plt.plot(x_axis_data, y_axis_data, 'g*--', alpha=0.5, linewidth=1, label='frecuency')#'bo-'表示蓝色实线，数据点实心原点标注
## plot中参数的含义分别是横轴值，纵轴值，线的形状（'s'方块,'o'实心圆点，'*'五角星   ...，颜色，透明度,线的宽度和标签 ，

plt.legend()  #显示上面的label
plt.xlabel('words') #x_label
plt.ylabel('number')#y_label
 
#plt.ylim(-1,1)#仅设置y轴坐标范围
plt.show()

for key, value in A:
    print(key,value)
```
